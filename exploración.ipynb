```python
from IPython import get_ipython
from IPython.display import display
import re
from google.colab import files
import pandas as pd
from google.colab import drive
import numpy as np
import kagglehub
from kagglehub import KaggleDatasetAdapter
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from sklearn.pipeline import Pipeline
from sklearn.model_selection import learning_curve
import dill

file_path = 'https://drive.google.com/uc?export=download&id=1Tt6OFJeC5s2D7rc88lOs7YD0_8Ij1oC5'
data_social = pd.read_csv(file_path)

#Coloca data válida donde hay datos inválidos
total_celdas = data_social.shape[0] * data_social.shape[1]
total_nulos = data_social.isnull().sum().sum()
porcentaje_nulos_total = (total_nulos / total_celdas) * 100

print(f"Porcentaje de datos nulos o vacíos: {porcentaje_nulos_total:.2f}%")
#Se aplica inyección de datos nulos al numero faltante
n_missing = int(total_celdas * 0.0513)

# Elige aleatoriamente las posiciones donde se colocarán los NaN
for _ in range(n_missing):
    i = np.random.randint(0, data_social.shape[0])  # fila aleatoria
    j = np.random.randint(0, data_social.shape[1])  # columna aleatoria
    data_social.iat[i, j] = np.nan  # insertar NaN

print(f"Porcentaje de datos nulos o vacíos: {porcentaje_nulos_total:.2f}%")

print("5 primeras filas")
print(data_social.head())
print("Columnas y tipos de datos")
print(data_social.info())
print("Número de filas y columnas")
print(data_social.shape)
print("Análisis previo de la información")
for num, name in enumerate(data_social.columns):
    if pd.api.types.is_numeric_dtype(data_social[name]):
      print(f"                      {num +1}. {name}")
      print(f"Media: {data_social[name].mean()}")
      print(f"Mediana: {data_social[name].median()}")
      print(f"Desviación estándar: {data_social[name].std()}")

print("Datos nulos")
print(data_social.isnull().sum())
print("Diferencia con respecto al género")
print(data_social['gender'].value_counts())
print("Diferencia con respecto al job type")
print(data_social['job_type'].value_counts())
print("Diferencia con respecto al social_platform_preference")
print(data_social['social_platform_preference'].value_counts())
print("Diferencia con respecto al gender, job_type, daily_social_media_time")
pd.set_option('display.max_rows', 100)
print(data_social[['gender', 'job_type', 'stress_level']].value_counts())


#se agrupan los datos de daily_social_media_time en categorias
#donde 3 es la más mala ya que son más de 5 horas al día
def assign_category(time):
  if time > 5:
    return 3
  elif (time > 3 and time <= 5):
    return 2
  else:
    return 1

data_social['category_social_m_time'] = data_social['daily_social_media_time'].apply(assign_category)
data_social.head()

print(data_social[['gender','category_social_m_time']].value_counts())
#Cuenta aplicando filtro con las peores categorias
filtered_data_social = data_social[data_social['category_social_m_time'].isin([2, 3]) ]
print(filtered_data_social[[ 'category_social_m_time',  'social_platform_preference']].value_counts())

print(data_social[[ 'category_social_m_time', 'actual_productivity_score']].value_counts())
print(data_social.info)
columnas_numericas = ['daily_social_media_time', 'actual_productivity_score',
                      'number_of_notifications', 'age', 'perceived_productivity_score',
                      'stress_level', 'screen_time_before_sleep']
data_social_numericas = data_social[columnas_numericas]

# Calcular la matriz de correlación
matriz_correlacion = data_social_numericas.corr()

# Visualizar la matriz de correlación
plt.figure(figsize=(10, 8))
sns.heatmap(matriz_correlacion, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriz de Correlación entre Variables')
plt.show()
#se visualiza una alta correlación entre 'perceived_productivity_score' y'actual_productivity_score'
variables_numericas = ['daily_social_media_time', 'stress_level', 'actual_productivity_score', 'number_of_notifications', 'age']
corr_matrix = data_social[variables_numericas].corr()
print(corr_matrix)

variables_numericas = ['perceived_productivity_score', 'actual_productivity_score']
corr_matrix = data_social[variables_numericas].corr()
print(corr_matrix)
data_social.groupby('category_social_m_time')[['stress_level', 'work_hours_per_day']].mean()


for name in data_social.columns:
  # Crear dos subpilots en uno, con respecto a category_social_m_time
  fig, axes = plt.subplots(1, 2, figsize=(12, 6))

  # Define el primer gráfico
  sns.boxplot(data=data_social, y=f'{name}', x='category_social_m_time', ax=axes[0])
  axes[0].set_title(f"Distribución de {name} (Boxplot)")
  axes[0].set_xlabel("Category Social Media Time")
  axes[0].set_ylabel(f'{name}')

  # Define el segundo gráfico
  sns.histplot(data=data_social, x=f'{name}', hue='category_social_m_time', kde=True, bins=20, palette="Set1", element="step", ax=axes[1])
  axes[1].set_title(f"Histograma de {name}")
  axes[1].set_xlabel(f'{name}')
  axes[1].set_ylabel("Frecuencia")
  axes[1].legend(title="Category Social Media Time, 1: (< 3h) 2: (3h-5h) 3: (> 5h)")

  plt.tight_layout()
  plt.show()

```
